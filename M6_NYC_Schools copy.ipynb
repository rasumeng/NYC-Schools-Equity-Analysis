**Outliers and Specialized Schools:** The histogram reveals notable outliers on the right side. Specifically, 30 schools (approximately 1.6% of all schools) exceed 2,000 students, with the largest serving 5,592 students. These outliers likely reflect specialized institutions, schools in high-density urban areas, or potential differences in grade configurations. The rarity of these extreme sizes (only 1.6% of schools) actually supports the finding that most NYC schools operate in the mid-sized range.

**Implications for Specialized Programs:** While enrollment size alone doesn't indicate school type, the presence of very large schools (2,000+) is notable. However, outliers could reflect:
- Geographic factors (borough density differences)
- Specialized programs serving students with disabilities

The data alone cannot definitively explain these outliers, but they indicate that NYC employs diverse school models beyond the standard mid-sized comprehensive high school.

# Part 3: Investigating Relationshipsâ€”Poverty and English Language Learners

## Hypothesis: Do Poverty and ESL Populations Overlap?

While analyzing individual distributions is valuable, identifying **relationships between variables** reveals deeper patterns. A key question in educational equity is whether socioeconomic disadvantage and language barriers tend to co-occur in schools, creating compounded challenges for both students and educators.

I hypothesized that schools with higher poverty rates would also tend to serve higher proportions of English learners. This would suggest that poverty and language barriers often occur together, requiring schools to simultaneously address both challenges.

To visualize this relationship, I created a scatter plot with poverty on the x-axis and English learner percentage on the y-axis.

```python
# Scatter plot: Poverty vs. English Learners
fig = px.scatter(
    df,
    x='percent_poverty',
    y='percent_english_learners',
    title='Relationship Between Poverty Rates and English Learner Populations',
    labels={
        'percent_poverty': 'Percent of Students in Poverty',
        'percent_english_learners': 'Percent of English Learners'
    },
    opacity=0.6,
    trendline='ols'  # Add a trend line to visualize relationship
)

fig.show()
```

### Key Findings: Poverty-ESL Relationship

**Co-occurrence Pattern:** The scatter plot shows that schools with higher poverty rates tend to also have higher ESL populations. This co-occurrence is meaningful from a resource-planning perspective: students experiencing both economic disadvantage and language barriers often attend the same schools. However, it is essential to note that **correlation does not imply causation**. Poverty does not cause ESL status (or vice versa); rather, they tend to co-occur due to underlying factors such as immigration patterns, geographic settlement, labor market dynamics, and demographic change.

**Positive Relationship Observed:** The scatter plot shows a positive relationship between poverty rates and English learner percentages. The trend line confirms that schools with higher poverty rates tend to also serve higher proportions of English learners. However, this is a *correlation*, not a causal relationship. The relationship is moderate (r = 0.32), meaning poverty explains only about 10% of the variance in ESL populationsâ€”other factors substantially influence the distribution of ESL students across schools.

**Why This Mattersâ€”Compounded Challenges:** The co-occurrence of poverty and ESL status means many students face **intersecting disadvantages simultaneously**. These students experience both economic hardship and language barriers to academic success. Schools serving high proportions of these students face resource challenges on multiple fronts.

**Implications for Educational Equity and Policy:** Schools with high concentrations of both poverty and ESL students require:
- **Enhanced ESL instructional capacity** beyond standard programming
- **Academic support services** that address both language development and content learning
- **Wraparound services** addressing poverty-related barriers (food security, transportation, mental health support)
- **Teacher professional development** in culturally responsive teaching and ESL methodology
- **Resource allocation** that prioritizes these high-need schools

The relationship between poverty and ESL demonstrates that effective education policy must address these factors holistically rather than in isolation.

# Part 4: Quantifying Relationshipsâ€”Correlation Analysis

## Moving Beyond Visual Inspection

Visual analysis provides intuitive understanding, but **quantifying relationships through correlation** enables more rigorous comparison. Correlation coefficients measure the strength and direction of linear relationships between variables, allowing me to identify which factors are most strongly connected in this dataset.

Correlation coefficients range from -1 to 1:
- **Correlation near 1**: Strong positive relationship (as one variable increases, so does the other)
- **Correlation near 0**: No linear relationship
- **Correlation near -1**: Strong negative relationship (as one increases, the other decreases)

**Interpretation Guidelines:**
- 0 to Â±0.2: Negligible/very weak
- Â±0.2 to Â±0.4: Weak to moderate  
- Â±0.4 to Â±0.6: Moderate to strong
- Â±0.6 to Â±0.8: Strong
- Â±0.8 to Â±1: Very strong

By calculating the correlation matrix, I can systematically identify which school characteristics are most strongly related.

## Calculating Correlations Across All Variables

```python
import numpy as np
# Calculate the correlation matrix (numeric columns only)
corr_matrix = df.select_dtypes(include=[np.number]).corr()

# Display the correlation matrix
print(corr_matrix)

# Replace diagonal values with 0 to ignore perfect self-correlations
np.fill_diagonal(corr_matrix.values, 0)

# Find the pair with the highest absolute correlation
max_corr = corr_matrix.abs().max().max()
max_pair = [(row, col) for row in corr_matrix.index for col in corr_matrix.columns 
            if abs(corr_matrix.loc[row, col]) == max_corr]

print(f"\nThe strongest correlation is {max_corr:.4f} between {max_pair[0][0]} and {max_pair[0][1]}.")
```

## Correlation Analysis Results

**Strongest Correlation Found: Percent English Learners & Percent Poverty = 0.3155**

**Interpretation:** The correlation of 0.32 represents a **weak-to-moderate positive relationship**. While statistically significant (given the large sample size of 1,834 schools), this correlation explains only about 10% of the variance in ESL populations (rÂ² â‰ˆ 0.10). This means poverty predicts only a small fraction of the variation in ESL distributions; other unmeasured factors drive much of the pattern.

**What This Tells Us:**
1. **Poverty and ESL co-occur more than random chance would predict** â€” the relationship is real and meaningful for resource planning
2. **But the relationship is far from deterministic** â€” knowing a school's poverty rate only weakly predicts its ESL population. Many factors influence ESL concentration independently
3. **The relationship is correlational, not causal** â€” this data does not establish that poverty causes higher ESL enrollment or vice versa. Geographic, demographic, and institutional factors likely drive both variables independently

**Absence of Strong Correlations:** No correlations exceed 0.33, suggesting that **school characteristics are relatively independent**. School size doesn't reliably predict disability rates, for example. This indicates that NYC's schools vary across multiple independent dimensions, not along a single "hierarchy of need." Each school requires individualized assessment rather than categorization by a single metric.

**Practical Implication:** Schools cannot be effectively categorized by a single need metric. Comprehensive support requires analyzing each school's unique profile across multiple dimensions (enrollment, disabilities, ESL, poverty) rather than ranking schools by one indicator.

# Conclusion: Key Takeaways for NYC Educational Equity

This analysis of NYC public high school data reveals several critical insights about the system's structure and challenges:

## 1. **A High-Need System**
NYC's public high schools serve a predominantly economically disadvantaged population (mean poverty 75%, median 79.7%). Poverty and language barriers *co-occur* in many schools, creating compounded resource demands. However, this analysis identifies correlation, not causation. Schools serving high concentrations of both low-income and ESL students face real operational challenges, and **evidence-based interventions from educational research would be needed** to determine the most effective support strategies. Plausible areas for investment include ESL instructional capacity, wraparound services, and teacher professional development, but effectiveness must be validated through rigorous evaluation.

School characteristics vary relatively independently (highest correlation = 0.32). This means:
- Each school requires individualized needs assessment
- Resource allocation cannot rely on single metrics

This analysis provides a snapshot of demographic distributions but cannot answer causal questions such as:
- *Why* do poverty and ESL co-occur in specific neighborhoods?

- Which interventions most effectively support high-need students?

Longitudinal analysis, qualitative research, and outcomes data would be essential to address these deeper questions.

## 6. **The Need for Targeted, Evidence-Based Investment**
Schools with high concentrations of multiple challenges (poverty + ESL + disabilities) likely represent areas of greatest need. However, resource allocation should be informed by:
- Outcomes data (not just demographic indicators)
- School-level needs assessments
- Evidence on intervention effectiveness
- Community and educator input
